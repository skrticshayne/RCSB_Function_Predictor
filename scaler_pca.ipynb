```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import joblib
import pandas as pd

# Load the raw dataset
file_path = '/Users/carmenshero/Desktop/Crumble_Cookie/TOTAL_DATASET_READY_TO_TRAIN.csv'
raw_data = pd.read_csv(file_path)

# Step 1: Define the columns for standardization and PCA
columns_to_standardize = list(raw_data.columns[1:18])  # B-S
columns_for_pca = list(raw_data.columns[19:22]) + list(raw_data.columns[22:-3])  # T-NV

# Step 2: Create a ColumnTransformer to handle both transformations
preprocessor = ColumnTransformer(transformers=[
    ('scaler', StandardScaler(), columns_to_standardize),  # Standardize columns B-S
    ('pca', PCA(n_components=2), columns_for_pca)          # PCA on columns T-NV
])

# Step 3: Wrap the ColumnTransformer in a Pipeline (optional, for extensibility)
scaler_pca_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor)  # Combine preprocessing steps
])

# Step 4: Fit the pipeline on the entire dataset
scaler_pca_pipeline.fit(raw_data)

# Step 5: Save the pipeline
pipeline_path = '/Users/carmenshero/Desktop/Crumble_Cookie/scaler_pca.pkl'
joblib.dump(scaler_pca_pipeline, pipeline_path)
print(f"Combined scaler and PCA pipeline saved successfully to: {pipeline_path}")
```
